## Customizando as visualizações

```python
import datetime
```

```python
fig = plt.figure(figsize = (24, 10))
eixo = fig.add_axes([0, 0, 1 ,1])
eixo.plot(dados['data'], dados['temperatura'], color  = 'darkblue')
#limitar o período do gráfico (fazer um 'recorte temporal')
eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo.set_title('Temperatura em 2014', fontsize = 25)
eixo.set_ylabel('°K', fontsize = 20)


'''
Parâmetros do  .plot de customização:

 lw = 04 ou 3 (aumenta/diminui a espessura da linha)
 ls = 'dotted' (pontilhado)
 marker = '0', '*', etc (adiciona marcadores as linhas)
 .grid(True) (adiciona uma grade no gráfico)
 pad = 20 (o titulo ficar um pouco acima do gráfico)


'''
```

```python
#fazendo 'visualização dentro de visualização'
#acrescentamos outro eixo na mesma figura

fig = plt.figure(figsize = (24, 10))
eixo = fig.add_axes([0, 0, 1 ,1])
eixo2 = fig.add_axes([0.6, 0.6, .4 , .4])

```

```python
#Fazendo um recorte/destaque do mês de maio


fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0, 0, 1 ,1])
eixo2 = fig.add_axes([0.6, 0.6, .4 , .4])

eixo.plot(dados['data'], dados['temperatura'], color  = 'g')
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1))
eixo.set_ylim(250, 340)
eixo.set_title('Temperatura em Maio/2014', fontsize = 25, pad = 15)
eixo.set_ylabel('°K', fontsize = 20)

#destacando o mês de maio
limite_inf = dados['data'] < datetime.datetime(2014, 5, 1)
limite_sup = dados['data'] > datetime.datetime(2014, 6, 1)

eixo2.plot(dados['data'], dados['temperatura'], color  = 'g')
eixo2.plot(dados[limite_inf]['data'], dados[limite_inf]['temperatura'], color  = 'b')
eixo2.plot(dados[limite_sup]['data'], dados[limite_sup]['temperatura'], color  = 'b')
eixo2.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo2.set_title('Temperatura em 2014', fontsize = 15)
eixo2.set_ylabel('°K', fontsize = 10)

```

```python
# identificar o máximo e o mínimo
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0, 0, 1 ,1])
eixo.plot(dados['data'], dados['temperatura'], color  = 'g')
eixo.set_title('Temperatura em 5 anos', fontsize = 25, pad = 15)

#criando as linhas limites e indicando o máximo e mínimo

eixo.axhline(max(dados['temperatura']), color = 'k', linestyle = '--')
eixo.axhline(min(dados['temperatura']), color = 'k', linestyle = '--')

x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados['temperatura'])
x2 = dados['data'][dados['temperatura'].idxmax() - 7000]
y2 = max(dados['temperatura'] - 5)

eixo.annotate("Máximo", xy = (x1, y1), fontsize = 20, xytext = (x2, y2), arrowprops = dict(facecolor = 'k'))

x1 = dados['data'][dados['temperatura'].idxmin()]
y1 = min(dados['temperatura'])
x2 = dados['data'][dados['temperatura'].idxmin() - 7000]
y2 = min(dados['temperatura'] + 5)

eixo.annotate("Mínimo", xy = (x1, y1), fontsize = 20, xytext = (x2, y2), arrowprops = dict(facecolor = 'k'))


```

## Algumas outras visualizações

```python
# ver o comportamento da temperatura por dia da semana

temperatura_dia_da_semana = dados.groupby('dia_da_semana')['temperatura'].mean()
#ordenando os dias
dias = ['Domingo', 'Segunda','Terça', 'Quarta', 'Quinta' ,'Sexta', 'Sábado']
temperatura_dia_da_semana = temperatura_dia_da_semana[dias]
temperatura_dia_da_semana
```

```python
# só para exemplificar, esses gráficos não necessariamente são a melhor forma de representar os dados abordados


# gráfico de barras
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

indice = range(len(temperatura_dia_da_semana))
cores = ['k', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temperatura_dia_da_semana, color = cores, edgecolor = 'k')
eixo.set_title('Temperatura média por dia da semana', fontsize = 15, pad = 10)
eixo.set_xlabel('Dia da semana', fontsize = 15)
eixo.set_xticks(indice)
eixo.set_xticklabels(dias)

eixo.set_ylabel('°K', fontsize = 15)
```

```python
# gráfico de pizza
fig = plt.figure(figsize=(5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

destacar = [.1, 0,0,0,0, .1, .1]

eixo.pie(temperatura_dia_da_semana, labels = temperatura_dia_da_semana.index, autopct= '%.1f%%', explode= destacar, shadow= True)
eixo.set_title('Temperatura média por dia da semana', fontsize = 15, pad = 10)
```

## Utilizando gráficos de Dispersão

```python
import numpy as np
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Dispersão', fontsize = 25, pad = 15)

eixo.set_xlabel('Comprimento da Sépala', fontsize = 15)
eixo.set_ylabel('Largura da Sépala', fontsize = 15)

eixo.tick_params(labelsize = 15) # Aumentar/diminuir o tamanho dos números dos eixos
 
```

```python
# Separando cores diferentes para espécies diferentes
cores = {
    'Iris-setosa': 'r',
    'Iris-versicolor': 'b',
    'Iris-virginica': 'g'
}

marcadores = {
    'Iris-setosa': 'x',
    'Iris-versicolor': 'o',
    'Iris-virginica': 'v'
}

fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])


for especie in df['espécie'].unique():
    tmp = df[df['espécie'] == especie]
    eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'], color = cores[especie], marker= marcadores[especie], s = 100)

#eixo.scatter(df['comprimento_sépala'], df['largura_sépala'])
eixo.set_title('Dispersão', fontsize = 25, pad = 15)

eixo.set_xlabel('Comprimento da Sépala', fontsize = 15)
eixo.set_ylabel('Largura da Sépala', fontsize = 15)

eixo.tick_params(labelsize = 15) # Aumentar/diminuir o tamanho dos números dos eixos

eixo.legend(cores, fontsize= 20)
```

```python
# Verificando a distribuição dos dados com boxplot

fig = plt.figure(figsize=(10, 6))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.boxplot(df['largura_pétala'])
eixo.set_title('Gráfico de caixa', fontsize = 15, pad = 10)
eixo.set_xticklabels(['largura_pétala'])

```

```python
df.columns
```

```python
# Mostrando todas as variáveis
fig = plt.figure(figsize=(15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

cores = ['r', 'b', 'g', 'orange']


caixas = eixo.boxplot(df.drop('espécie', axis = 1).values, patch_artist= True)
eixo.set_title('Gráfico de caixa', fontsize = 15, pad = 10)
eixo.set_xticklabels(df.drop('espécie', axis = 1).columns)

for caixa, cor in zip(caixas['boxes'], cores): # no matplotlib para preencher cada caixa com uma cor (é melhor se puder usar o seaborn)
    caixa.set(color = cor)

for outlier in caixas['fliers']:
    outlier.set(marker = 'x', markersize = 10)

```

```python
# Histogramas

fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.hist(df['comprimento_pétala'], bins = 20, density = True)
eixo.set_title('Histograma', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da Sépala', fontsize = 15)
```

```python
fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins = 20)
eixo.set_title('Histograma', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da Sépala', fontsize = 15)


# anotando as informações de média e desvio
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
            xy = (5, 25), fontsize = 25)

#linhas indicando a média e mediana

eixo.axvline(mu, color = 'k', linestyle = '--')
eixo.annotate('Média', xy = (mu - .5, 28), fontsize = 20)

mediana = df['comprimento_pétala'].median()
eixo.axvline(mediana, color = 'g', linestyle = '--')
eixo.annotate('Mediana', xy = ( mediana + .05, 33), fontsize = 20)
```

```python
# dividindo por espécie


# iris versicolor
df_iv = df[df['espécie'] == 'Iris-versicolor']

fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])

mu, sigma = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'], bins = 20)
eixo.set_title('Iris Versicolor', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da Sépala', fontsize = 15)


# anotando as informações de média e desvio
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
            xy = (4.8, 6), fontsize = 25)

#linhas indicando a média e mediana

eixo.axvline(mu, color = 'k', linestyle = '--')
eixo.annotate('Média', xy = (mu - .15, 6), fontsize = 15)

mediana = df_iv['comprimento_pétala'].median()
eixo.axvline(mediana, color = 'g', linestyle = '--')
eixo.annotate('Mediana', xy = ( mediana + .02, 7.1), fontsize = 15, color = 'g')

#salvar a figura
fig.savefig('histograma_iv.png', bbox_inches = 'tight')

```

```python
# iris setosa
df_is = df[df['espécie'] == 'Iris-setosa']

fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])

mu, sigma = df_is['comprimento_pétala'].mean(), df_is['comprimento_pétala'].std()

eixo.hist(df_is['comprimento_pétala'], bins = 20)
eixo.set_title('Iris Setosa', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da Sépala', fontsize = 15)


# anotando as informações de média e desvio
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
            xy = (1.8, 12), fontsize = 25)

#linhas indicando a média e mediana

eixo.axvline(mu, color = 'k', linestyle = '--')
eixo.annotate('Média', xy = (mu - .06, 13), fontsize = 15)

mediana = df_is['comprimento_pétala'].median()
eixo.axvline(mediana, color = 'g', linestyle = '--')
eixo.annotate('Mediana', xy = ( mediana + 0.02, 14.1), fontsize = 15, color = 'g')

#salvar a figura
fig.savefig('histograma_is.png', bbox_inches = 'tight')
```

```python
# iris virginica
df_ivc = df[df['espécie'] == 'Iris-virginica']

fig = plt.figure(figsize=(12, 6))
eixo = fig.add_axes([0, 0, 1, 1])

mu, sigma = df_ivc['comprimento_pétala'].mean(), df_ivc['comprimento_pétala'].std()

eixo.hist(df_ivc['comprimento_pétala'], bins = 20)
eixo.set_title('Iris Virginica', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da Sépala', fontsize = 15)


# anotando as informações de média e desvio
eixo.annotate('$\mu = {0:.2f}$\n$\sigma = {1:.2f}$'.format(mu, sigma),
            xy = (1.8, 12), fontsize = 25)

#linhas indicando a média e mediana

eixo.axvline(mu, color = 'k', linestyle = '--')
eixo.annotate('Média', xy = (mu + .06, 8.5), fontsize = 15)

mediana = df_ivc['comprimento_pétala'].median()
eixo.axvline(mediana, color = 'g', linestyle = '--')
eixo.annotate('Mediana', xy = ( mediana + 0.02, 8), fontsize = 15, color = 'g')

#salvar a figura
fig.savefig('histograma_ivc.png', bbox_inches = 'tight')
```
