```python
deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
recovered = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
display(deaths.head())
display(recovered.head())
```

```python
recovered.groupby('Country/Region').sum()[last_date].head()
```

```python

# funÃ§Ã£o que retorna os casos mais recentes por paÃ­s
def latest_by_country(data):
    return data.groupby('Country/Region').sum().iloc[:, -1]
```

```python
latest_by_country(recovered)
```

```python
informations = [latest_by_country(dados), latest_by_country(deaths), latest_by_country(recovered)]
```

```python
combinados = pd.concat(informations, axis = 1)
combinados.columns = ['confirmados', 'mortes', 'recuperados']
combinados
```

```python
# taxa de letalidade expressa em % 

# uma forma simples: mortes / confirmados
somas = combinados.sum()
taxa_letalidade1 = somas['mortes'] / somas['confirmados'] 
print('{}%'.format((taxa_letalidade1 * 100).round(2)))
```

```python
letalidade_por_pais = combinados['mortes'] / combinados['confirmados'] * 100
combinados['letalidade_por_pais'] = letalidade_por_pais
```

```python
combinados.sort_values('letalidade_por_pais', ascending=False).head()
```

```python
import seaborn as sns

sns.scatterplot(data = combinados.query("confirmados> 1000"), x = 'confirmados', y= 'letalidade_por_pais')
```

```python
sns.distplot(combinados.query("confirmados > 40")['letalidade_por_pais'])
sns.set(rc={'figure.figsize':(20, 6)})
```

## Separando o Brasil no Dataset

```python
is_brazil = dados['Country/Region'] == 'Brazil'
```

```python
dados['is_brazil'] = is_brazil
dados.sample(10)
```

```python
dados.query("is_brazil")
```

```python
# definindo o nome onde tem True como Brazil e false como Others
import numpy as np
dados['is_brazil_labels'] = np.where(dados['is_brazil'], 'Brazil', 'Others')
```

```python
casos = dados.groupby("is_brazil_labels").sum()[last_date].loc['Brazil']

populacao_estimada_br = 213300000

#incidÃªncia covid por 100 mil habitantes - Brasil
incidence = casos / populacao_estimada_br * 100000
incidence.round()
```

```python
dados.groupby("is_brazil_labels").sum()[last_date].plot(kind = 'barh', color = 'r')
plt.title(f'Casos acumulados atÃ© {last_date}', fontsize = 20)
```
