# Removendo Outliers 

```python
 dados2 = pd.read_csv('datasets/aluguel_amostra.csv', sep=';')
dados2.head()
```

```python
Q1 = dados2['Valor m2'].quantile(.25)
Q3 = dados2['Valor m2'].quantile(.75)
IIQ = Q3-Q1
limite_if = Q1 - 1.5 * IIQ
limite_sup = Q3 + 1.5 * IIQ
```

```python
print('Q1 = {}' .format(Q1))
print('Q3 = {}' .format(Q3))
print('IIQ = {}' .format(IIQ))    
print('Limite inferior = {}' .format(limite_if))       
print('Limite superior = {}' .format(limite_sup))       
```

```python
dados2.boxplot(['Valor'], by = ['Tipo'])
```

```python
grupo_tipo = dados2.groupby('Tipo')['Valor']
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25)
Q3 = grupo_tipo.quantile(.75)
IIQ = Q3-Q1
limite_if = Q1 - 1.5 * IIQ
limite_sup = Q3 + 1.5 * IIQ
```

```python
Q1
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados2['Tipo'] == tipo
    ta_no_limite = (dados2['Valor'] >= limite_if[tipo]) & (dados2['Valor'] <= limite_sup[tipo])
    selecao = eh_tipo & ta_no_limite
    dados_selecao = dados2[selecao]
    dados_new = pd.concat([dados_new, dados_selecao])
```

```python
dados_new.boxplot(['Valor'], by = ['Tipo'])
```

```python
dados_new.to_csv('datasets/aluguel_residencial_sem_outliers.csv', sep = ';', index = False)
```
